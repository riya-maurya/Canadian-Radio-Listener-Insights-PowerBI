## Data Cleaning

```python
columns_to_drop = ['DGUID','UOM_ID','SCALAR_ID','SCALAR_FACTOR','VECTOR','COORDINATE']
existing_columns = [col for col in columns_to_drop if col in df.columns]
df = df.drop(existing_columns, axis=1)
```

```python
df
```

```python
df_cleaned = df.rename(columns={
    'REF_DATE': 'Year',
    'Radio station format': 'Genre',
    'Audience occupation': 'Occupation'
})
```

```python
print(df_cleaned.isnull().sum())

# Preview cleaned data
df_cleaned.head()
```

```python
import matplotlib.pyplot as plt

# Group by Genre and get mean VALUE
genre_avg = df_cleaned.groupby("Genre")["VALUE"].mean().sort_values(ascending=False)

# Display the table
print(genre_avg)

# Plot bar chart
plt.figure(figsize=(12,6))
genre_avg.plot(kind='bar', color='skyblue')
plt.title("Average Popularity of Radio Genres (All Years)")
plt.ylabel("Average % of Listeners")
plt.xlabel("Radio Genre")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by Occupation and get average listener percentage
occupation_avg = df_cleaned.groupby("Occupation")["VALUE"].mean().sort_values(ascending=False)

# Show top occupations
print(occupation_avg)

# Plot bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
occupation_avg.plot(kind='bar', color='lightgreen')
plt.title("Average Radio Listening by Occupation")
plt.ylabel("Average % of Listeners")
plt.xlabel("Occupation")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by Occupation and Genre, calculate mean listenership
grouped = df_cleaned.groupby(["Occupation", "Genre"])["VALUE"].mean().reset_index()

# For each Occupation, get top 3 genres
top3_genres_by_job = grouped.groupby("Occupation").apply(
    lambda x: x.sort_values("VALUE", ascending=False).head(3)
).reset_index(drop=True)

# Display results
top3_genres_by_job
```

```python
# Ensure Year is numeric
df_cleaned['Year'] = pd.to_numeric(df_cleaned['Year'], errors='coerce')
```

```python
# Filter only News and Jazz
# Check available genres first
available_genres = df_cleaned['Genre'].unique()
print("Available genres:", available_genres)

```

```python
# Ensure Year is numeric
df_cleaned['Year'] = pd.to_numeric(df_cleaned['Year'], errors='coerce')

# Use updated genres
selected_genres = ['Talk radio format', 'Dance music']
trend_df = df_cleaned[df_cleaned['Genre'].isin(selected_genres)]

# Group by Year and Genre
trend_grouped = trend_df.groupby(['Year', 'Genre'])['VALUE'].mean().reset_index()

# Pivot for plotting
trend_pivot = trend_grouped.pivot(index='Year', columns='Genre', values='VALUE')

# Plot line chart
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
trend_pivot.plot(marker='o', linewidth=2)
plt.title("Trend of Talk Radio vs Dance Music (1999â€“Present)")
plt.ylabel("Average % of Listeners")
plt.xlabel("Year")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by Occupation + Genre and calculate average listener %
combo_avg = df_cleaned.groupby(["Occupation", "Genre"])["VALUE"].mean().reset_index()

# Sort by highest % of listeners
top3_combos = combo_avg.sort_values(by="VALUE", ascending=False).head(3)

# Display result
top3_combos
```
